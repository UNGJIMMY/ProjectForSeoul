```python
Insight

비수도권의 인구 증감 지표가 모두 긍정적이었음에도 불구하고 인구 비율이 부정적이었으므로, 
인구 비율이 긍정적인 수도권의 인구 증감 지표에서 유일하게 긍정적인 생산성과,
인구 수용 비용의 절대적인 수치를 주목해야 한다.
```

```python
수도권/비수도권 인구비율 확인

2016~2025. "행정동별 주민등록 인구 및 세대현황". 행정안전부.

수도권(수도권 지하철 연결 도시): 서울, 인천, 경기도 
```

```python
import pandas as pd

filePath = "../../temp_data/0_raw/2016-2025_한국인구.csv"
df = pd.read_csv(filePath, encoding='cp949', thousands=',')

print(df.head())
print(df.columns)
```

```python
# 필요 행(전국/수도권/비수도권), 칼럼(행정구역/각 연도별 총 인구수)만 추출

df = df[df['행정구역'].str.contains('전국|서울특별시|인천광역시|경기도', na=False)]

# 수도권

# 칼럼 추출
pop_cols = [col for col in df.columns if '총인구수' in col]

seoul    = df[df['행정구역'].str.contains('서울')][pop_cols].iloc[0]
incheon  = df[df['행정구역'].str.contains('인천')][pop_cols].iloc[0]
gyeonggi = df[df['행정구역'].str.contains('경기도')][pop_cols].iloc[0]

capital = seoul + incheon + gyeonggi

# 비수도권

national = df[df['행정구역'].str.contains('전국')][pop_cols].iloc[0]
non_capital = national - capital

# staging_df
staging_df = pd.DataFrame({
    '행정구역': ['전국', '수도권', '비수도권'],
    **{col.replace('년_총인구수', '').strip(): [national[col], capital[col], non_capital[col]] 
       for col in pop_cols}
})

# 백분율 전환 (가시성 위함, 전국 = 100.0 기준, 소수점 2자리)
years = [col.replace('년_총인구수', '').strip() for col in pop_cols]

for year in years:
    staging_df[year] = (staging_df[year] / staging_df.loc[staging_df['행정구역'] == '전국', year].values[0] * 100).round(2)

print(staging_df)
```

```python
# long data transformation

# 1. '행정구역' 컬럼을 인덱스로 설정합니다.
staging_df = staging_df.set_index('행정구역')

# 2. 데이터프레임을 전치(행/열 전환)합니다.
df_transposed = staging_df.transpose()

# 3. 인덱스 이름을 '연도'로 설정합니다.
df_transposed.index.name = '연도'

# 결과 확인
print(df_transposed)
```

```python
df_transposed.to_csv("../../temp_data/1_staging/staging_2016-2025_수도권인구비율.csv", encoding='utf-8-sig')
```

```python
import pandas as pd

increaseFactor_filePath = "../../temp_data/0_raw/한국도시인구증가요인_증감률.csv"
increaseFactor_df = pd.read_csv(increaseFactor_filePath, encoding='utf-8-sig')

seoul_filePath = "../../temp_data/1_staging/staging_2008-2019_수도권인구비율.csv"
seoul_df = pd.read_csv(seoul_filePath, encoding='utf-8-sig')

print(increaseFactor_df)
print(seoul_df)
```

```python
# 통합된 시각화를 위한 rows(2008, 2019), columns(수도권, 비수도권) 수정

staging_seoul_df = seoul_df.iloc[[0,11],[0,2,3]]

print(staging_seoul_df)
```

```python
# 연도/지역별 인구비율과 인구 증감 요인과의 상관관계를 확인하는 그래프
```

```python
# --- 데이터 전처리 및 병합 (핵심 단계) ---

# staging_seoul_df를 Long Format으로 변환 (지역을 컬럼에서 행 데이터로 내림)
staging_melted = staging_seoul_df.melt(
    id_vars='연도', 
    var_name='지역', 
    value_name='인구비율'
)

# 두 데이터프레임 병합 (연도와 지역을 기준으로)
merged_df = pd.merge(increaseFactor_df, staging_melted, on=['연도', '지역'])

# print("=== 병합된 데이터 미리보기 ===")
print(merged_df)
print("-" * 30)
```

```python
# --- 시각화: 상관관계 분석 ---

import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib.font_manager as fm

# --- 한글 폰트 설정 (환경에 맞게 자동 설정) ---
import platform
system_name = platform.system()
if system_name == 'Windows':
    plt.rc('font', family='Malgun Gothic')
elif system_name == 'Darwin': # Mac
    plt.rc('font', family='AppleGothic')
else: # Colab 등 Linux
    plt.rc('font', family='NanumGothic')

# plt.rcParams['font.family'] = 'Apple SD Gothic Neo'
plt.rcParams['axes.unicode_minus'] = False # 마이너스 기호 깨짐 방지


# --- 시각화: 지역별 서브플롯(Subplot) 그리기 ---
# 1행 2열의 그래프 공간 생성 (왼쪽: 수도권, 오른쪽: 비수도권)
fig, axes = plt.subplots(1, 2, figsize=(15, 6))

regions = ['수도권', '비수도권']
colors = {'생산성': 'red', '쾌적도': 'green', '인구수용비용': 'blue'}
markers = {'생산성': 'o', '쾌적도': '^', '인구수용비용': 's'}

for i, region in enumerate(regions):
    ax1 = axes[i]
    region_data = merged_df[merged_df['지역'] == region]
    
    # [좌측 Y축] 인구 비율 (막대 그래프 혹은 굵은 선) - 여기선 강조를 위해 영역 채우기 사용
    color_pop = 'gray'
    line1 = ax1.plot(region_data['연도'], region_data['인구비율'], color=color_pop, 
                     linestyle='--', marker='o', linewidth=3, markersize=10, label='인구비율(Left)')
    ax1.set_xlabel('연도', fontsize=12)
    ax1.set_ylabel('인구 비율 (%)', color=color_pop, fontsize=12)
    ax1.tick_params(axis='y', labelcolor=color_pop)
    ax1.set_title(f'{region} 인구비율 vs 증가요인 변화', fontsize=15, fontweight='bold')
    
    # X축 눈금을 데이터에 있는 연도로만 설정
    ax1.set_xticks(region_data['연도'])

    # [우측 Y축] 3가지 요인 (생산성/쾌적도/인구수용비용)
    ax2 = ax1.twinx() # x축을 공유하는 쌍둥이 축 생성
    
    lines = []
    for factor in ['생산성', '쾌적도', '인구수용비용']:
        l = ax2.plot(region_data['연도'], region_data[factor], 
                     color=colors[factor], marker=markers[factor], 
                     linestyle='-', linewidth=2, label=f'{factor}(Right)')
        lines.append(l[0])
        
    ax2.set_ylabel('요인 지수 (Index)', color='black', fontsize=12)
    
    # 범례 합치기 (좌측축 + 우측축)
    lns = line1 + lines
    labs = [l.get_label() for l in lns]
    ax1.legend(lns, labs, loc='upper center', bbox_to_anchor=(0.5, -0.15), ncol=2)

    # 그리드 추가
    ax1.grid(True, axis='x', linestyle=':', alpha=0.6)

plt.tight_layout()
plt.show()
```

```python
Insight

비수도권의 인구 증감 지표가 모두 긍정적이었음에도 불구하고 인구 비율이 부정적이었으므로, 
인구 비율이 긍정적인 수도권의 인구 증감 지표에서 유일하게 긍정적인 생산성과,
인구 수용 비용의 절대적인 수치를 주목해야 한다.

* 이 그래프를 올바르게 해석하기 위해 아래 사실을 유념해야한다.
    
1. 인구 증감 요인과 인구비율의 실제 조사기간을 다르고, 데이터 부족으로 인한 오차가 존재한다.
    (2005~2019)      (2008~2019)
2. 생산성/쾌적도는 인구 상승의 긍정적인 영향을, 인구수용비용은 부정적인 영향을 끼치므로, 올바른 음/양의 방향에서 해석해야 한다.
```
